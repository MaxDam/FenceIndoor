```python
import pandas as pd
import numpy as np
import json
import annEngine2 as ann
import matplotlib.pyplot as plt
%matplotlib inline
```

## inizializza le classi ed importa i dati

```python
getData = ann.GetData()
ae = ann.Autoencoder()
fc = ann.FullyConnectionLayer()
#X, Y = getData.getTrainingData()
X, Y = getData.getTrainingDataFromFile()
getData.plotInput(X)
getData.plotTrain(X,Y)
```

## addestra l'AE

```python
ae.verbose=0
scoreAE = ae.buildAndFit(X)
print('Test score:', scoreAE[0])
print('Test accuracy:', scoreAE[1])
ae.plotTrain()
```

## Addestra l'ANN

```python
fc.verbose=0
fc.use_dropout=False
X = ae.predict(X)
scoreANN = fc.buildAndFit(X, Y)
print('Test score:', scoreANN[0])
print('Test accuracy:', scoreANN[1])
fc.plotTrain()
fc.confusionMatrix(X,Y)
```
